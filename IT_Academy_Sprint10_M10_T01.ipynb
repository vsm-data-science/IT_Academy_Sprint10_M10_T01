## BeautifulSoup

### a) http://quotes.toscrape.com

### Hacemos la llamada a la web

```python
url_a = "http://quotes.toscrape.com"
```

### Se realiza una solicitud GET a la URL utilizando la función get de la biblioteca requests.

La respuesta HTTP se almacena en la variable r. Esta respuesta contiene la información recuperada del servidor web.

```python
r = requests.get(url_a)
r.status_code
```

El código de estado 200 indica que la solicitud fue exitosa.

### Se crea un objeto BeautifulSoup llamado soup a partir del contenido de la respuesta (r.content).

Este paso convierte el contenido HTML de la página web en una estructura de datos que se puede navegar y buscar fácilmente.

```python
soup = bs(r.content)
print(soup.prettify()) 
```

### Va a encontrar todos los elementos HTML "div" que tienen la clase CSS "quote" en la estructura de la página web.

```python
quote_elements = soup.find_all("div", class_="quote")
quote_elements
```

### Creamos un dataframe

Y extraemos la información específica (frases y autores) de cada elemento y la agregamos al DataFrame.

```python
data = []

for quote_element in quote_elements:
    quote = quote_element.find("span", class_="text")
    author = quote_element.find("small", class_="author")

    data.append({"Frase": quote.text.strip(),
                 "Autor": author.text.strip()})

quotes_df = pd.DataFrame(data)
```

```python
quotes_df
```

### Scrapy

```python
class OIVSpider(scrapy.Spider):
    name = 'oiv_spider'
    allowed_domains = ['www.oiv.int']
    start_urls = ['https://www.oiv.int/es']

    def parse(self, response):
        # Extraer títulos de noticias
        news_titles = response.css('div.views-field-title span.field-content a::text').extract()

        for title in news_titles:
            yield {
                'title': title.strip()
            }
```

Configuramos del proceso de Scrapy y ejecución

```python
from scrapy.crawler import CrawlerProcess
from scrapy.utils.project import get_project_settings
process = CrawlerProcess(get_project_settings())
process.crawl(OIVSpider)
process.start()
```
